```python
import os
opath = r'.\\pos'
ageFile = open("ageText.txt",'r')
count = 0
for line in ageFile:
    print(len(line),count,line)
    if (len(line)>0):
        docfile = open(os.path.join(opath,"doc_"+str(count)+".txt"),'w')
        docfile.write(line.rstrip())
        docfile.close()
    count += 1
        
```

```python
import os
import random
opath = r'.\\neg'
nonAgeFile = open("commoncore(CC)questions.txt",'r').readlines()
doclength = len(nonAgeFile)
print(doclength)
count = 0
index = random.sample(range(0,doclength),124)
lines = [nonAgeFile[ind].rstrip() for ind in index]
print(lines)
for line in lines:
   docfile = open(os.path.join(opath,"doc_"+str(count)+".txt"),'w')
   docfile.write(line)
   count += 1
   docfile.close()
      
```
